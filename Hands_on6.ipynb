```python
# 1️⃣ Group by OrderDate
sales_by_date = df.groupby("OrderDate")["Total"].sum()

plt.figure(figsize=(12,6))
sales_by_date.plot(kind="bar", color="skyblue")
plt.title("Total Sales per Order Date")
plt.xlabel("Order Date")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()

# 2️⃣ Group by Region
sales_by_region = df.groupby("Region")["Total"].sum()

plt.figure(figsize=(8,6))
sales_by_region.plot(kind="bar", color="lightgreen")
plt.title("Total Sales per Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()

# 3️⃣ Group by Rep
sales_by_rep = df.groupby("Rep")["Total"].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
sales_by_rep.plot(kind="bar", color="salmon")
plt.title("Total Sales per Representative")
plt.xlabel("Representative")
plt.ylabel("Total Sales")
plt.show()
```
